Question 1 : What is the difference between multithreading and multiprocessing?
- Definition:
  - Multithreading means running multiple threads within a single process.
  - Multiprocessing means running multiple independent processes simultaneously.
- Memory Usage:
  - Threads share the same memory space of one process.
  - Processes have separate memory spaces for each instance.
- Fault Tolerance:
  - If one thread crashes, it can affect the entire process.
  - If one process crashes, it does not affect other processes.
- Parallelism:
  - Multithreading provides concurrency but not always true parallelism (especially in languages with a Global Interpreter Lock like Python).
  - Multiprocessing provides true parallelism because each process runs on a different CPU core.
- Performance Focus:
  - Multithreading is better for I/O-bound tasks (like file I/O, network requests).
  - Multiprocessing is better for CPU-bound tasks (like data processing, computations).
- Examples:
  - In Java: Thread class for multithreading, multiple JVM processes for multiprocessing.
  - In Python: threading module for threads, multiprocessing module for processes.




Question 2: What are the challenges associated with memory management in Python?
- Python uses automatic garbage collection (mainly reference counting), which can make it harder to predict exactly when memory is freed.
- This lack of manual control can lead to unexpected memory usage in large applications.
- Large data structures (like big lists, dictionaries, or NumPy arrays) can consume a lot of memory if not carefully managed or deleted when no longer needed.
- The Global Interpreter Lock (GIL) limits true parallel memory access, which can affect performance in multithreaded memory-intensive applications.
- Python creates many small objects (like integers, strings), and frequent allocation/deallocation can increase overhead if not optimized.
- Python’s automatic memory management simplifies coding but can cause issues like memory leaks, fragmentation, and unpredictable garbage collection in large or long-running programs.

Question 3:Write a Python program that logs an error message to a log file when a division by zero exception occurs.


````python
import logging

# Configure logging settings
logging.basicConfig(
    filename='error.log',          # Log file name
    level=logging.ERROR,           # Log only error messages and above
    format='%(asctime)s - %(levelname)s - %(message)s'
)

try:
    num1 = int(input("Enter numerator: "))
    num2 = int(input("Enter denominator: "))
    result = num1 / num2
    print("Result:", result)

except ZeroDivisionError as e:
    logging.error("Division by zero error occurred: %s", e)
    print("Error: Cannot divide by zero. Check error.log for details.")
````

Question 4:Write a Python program that reads from one file and writes its content to another file.

````python
source_file = "source.txt"
destination_file = "destination.txt"

try:
    with open(source_file, 'r') as src:
        content = src.read()

    with open(destination_file, 'w') as dest:
        dest.write(content)

    print("File copied successfully!")

except FileNotFoundError:
    print("Error: Source file not found.")
except Exception as e:
    print("An error occurred:", e)
````

Question 6: What are the differences between NumPy arrays and Python lists?
- Data Type Consistency:
  - NumPy Array: All elements must be of the same data type.
  - Python List: Can hold mixed data types (ints, strings, objects, etc.).
- Performance:
  - NumPy Array is faster for numerical computations due to vectorized operations and optimized C backend.
  - Python List is slower for large numerical computations since operations are done element by element.
- Memory Efficiency:
  - NumPy Array uses less memory because it stores elements in a contiguous block of memory.
  - Python List is less memory-efficient; stores references to objects, not the raw data.
- Operations:
  - NumPy Array Supports vectorized operations (like array + 5 adds 5 to all elements).
  - Python Lists requires loops or list comprehensions for element-wise operations.
- Dimensions:
  - NumPy Array Can be multi-dimensional (1D, 2D, 3D, …).
  - Python List is Mainly 1D; nested lists can be used for multi-dimensional structures, but handling is less efficient.
- Functionality:
  - NumPy Array Comes with a rich set of mathematical, statistical, and linear algebra functions.
  - Python Lists have Limited built-in functions; most operations require manual coding or external libraries.
- Slicing & Indexing:
  - NumPy Arrays support Advanced slicing, boolean indexing, and broadcasting are supported.
  - Python Lists suport Basic slicing; no broadcasting or fancy indexing.
- NumPy arrays are homogeneous, faster, memory-efficient, and powerful for numerical operations, while Python lists are flexible, heterogeneous, and easier for general-purpose storage.

Question 7:Explain the difference between apply() and map() in Pandas
- Applicability:
  - map() → Mainly used with Pandas Series (one-dimensional data).
  - apply() → Can be used with both Series and DataFrame (1D or 2D).
- Functionality:
  - map() → Applies a function, dictionary, or Series element-wise on a Series.
  - apply() → Applies a function along an axis (rows or columns) of a DataFrame or element-wise on a Series.
- Return Type:
  - map() → Always returns a Series.
  - apply() → Can return a Series, DataFrame, or scalar, depending on the function and axis.
- Flexibility:
  - map() → Less flexible, best for simple element-wise transformations.
  - apply() → More flexible, can perform row-wise, column-wise, or complex operations.
- Example:
  ```py
  import pandas as pd

  df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

  # map() on Series
  df['A'] = df['A'].map(lambda x: x * 2)  # multiplies each element by 2

  # apply() on DataFrame
  df['Sum'] = df.apply(lambda row: row['A'] + row['B'], axis=1)  # sum row-wise
  ```

- Short Summary:
  - map() → element-wise operations on Series, simple and fast.
  - apply() → more general, works on Series and DataFrame, supports row/column operations.

Question 8: Create a histogram using Seaborn to visualize a distribution.

Question 10: Calculate the correlation matrix using Seaborn and visualize it with a heatmap

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Math': [85, 78, 90, 95, 70],
    'Physics': [80, 75, 88, 92, 65],
    'Chemistry': [78, 74, 85, 90, 60],
    'Biology': [82, 70, 89, 94, 68]
}

df = pd.DataFrame(data)

# Calculate correlation matrix
corr_matrix = df.corr()
print("Correlation Matrix:\n", corr_matrix)

# Visualize correlation matrix with heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
````
